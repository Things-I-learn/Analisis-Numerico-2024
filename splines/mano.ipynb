```python
import matplotlib
matplotlib.use('TkAgg')  # Asegura que el backend es interactivo

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 6))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Haz clic para marcar los puntos de la silueta de tu mano")

points = plt.ginput(n=-1, timeout=0)  

x, y = zip(*points)
ax.plot(x + (x[0],), y + (y[0],), 'bo-') 

plt.show()
```

```python
import numpy as np
np.save("datos_eguar.npy", points) 
array_cargado = np.load("datos_eguar.npy")
print(array_cargado)
```
