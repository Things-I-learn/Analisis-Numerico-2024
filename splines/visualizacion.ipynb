```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos
datos1 = np.load("datos_beimar.npy")  # Primer conjunto de datos
datos2 = np.load("datos_eguar.npy")    # Segundo conjunto de datos

# Extraer coordenadas x, y
x1, y1 = zip(*datos1)
x2, y2 = zip(*datos2)

# Crear la figura y los subgráficos en una sola fila
fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 fila, 2 columnas

# Graficar el primer conjunto de datos
axes[0].scatter(x1, y1, color="blue", alpha=0.7)
axes[0].set_title("Datos Beimar")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")

# Graficar el segundo conjunto de datos
axes[1].scatter(x2, y2, color="red", alpha=0.7)
axes[1].set_title("Datos Eduards")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import os
from glob import glob
from PIL import Image

# Obtener todas las imágenes
image_files = sorted(glob("spline_*.png"))
n_images = len(image_files)

# Definir filas y columnas
rows = 2
cols = (n_images + 1) // 2  # Ajustar columnas para dos filas

# Crear figura grande
fig, axes = plt.subplots(rows, cols, figsize=(20, 10))
axes = axes.flatten()

# Cargar y mostrar imágenes
titles = [os.path.basename(img) for img in image_files]
for i, img_file in enumerate(image_files):
    img = Image.open(img_file)
    axes[i].imshow(img)
    #axes[i].set_title(titles[i], fontsize=14)
    axes[i].axis("off")

# Desactivar ejes vacíos
for i in range(n_images, len(axes)):
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```
