```python
import numpy as np

# Definimos la función de ejemplo
def f(x):
    return np.sin(x)  # Puedes cambiar esta función por cualquier otra

# Derivada exacta para comparar (en este caso la derivada de sin(x) es cos(x))
def f_derivada_exacta(x):
    return np.cos(x)

# Parámetros
x0 = np.pi / 4  # Punto donde queremos evaluar la derivada
h = 1e-5  # Tamaño del paso

# Aproximaciones
# Diferencia Hacia Adelante
derivada_adelante = (f(x0 + h) - f(x0)) / h

# Diferencia Hacia Atrás
derivada_atras = (f(x0) - f(x0 - h)) / h

# Diferencia Central
derivada_central = (f(x0 + h) - f(x0 - h)) / (2 * h)

# Derivada Exacta para comparar
derivada_exacta = f_derivada_exacta(x0)

# Mostrar resultados
print("Aproximación por Diferencia Adelante: ", derivada_adelante)
print("Aproximación por Diferencia Atrás:    ", derivada_atras)
print("Aproximación por Diferencia Central:  ", derivada_central)
print("Derivada Exacta:                      ", derivada_exacta)

# Errores
print("Error con Diferencia Adelante: ", abs(derivada_exacta - derivada_adelante))
print("Error con Diferencia Atrás:    ", abs(derivada_exacta - derivada_atras))
print("Error con Diferencia Central:  ", abs(derivada_exacta - derivada_central))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función de ejemplo
def f(x):
    return np.sin(x)  # Puedes cambiar esta función

# Derivada exacta para comparar (en este caso la derivada de sin(x) es cos(x))
def f_derivada_exacta(x):
    return np.cos(x)

# Parámetro
x0 = np.pi / 4  # Punto donde queremos evaluar la derivada

# Valores de h decrecientes
h_values = np.logspace(-1, -15, 100)  # 100 valores de h entre 1e-1 y 1e-15

# Listas para almacenar los errores
error_adelante = []
error_atras = []
error_central = []

# Derivada exacta en x0
derivada_exacta = f_derivada_exacta(x0)

# Cálculo de la derivada y el error para cada valor de h
for h in h_values:
    # Diferencia hacia adelante
    derivada_adelante = (f(x0 + h) - f(x0)) / h
    error_adelante.append(abs(derivada_adelante - derivada_exacta))
    
    # Diferencia hacia atrás
    derivada_atras = (f(x0) - f(x0 - h)) / h
    error_atras.append(abs(derivada_atras - derivada_exacta))
    
    # Diferencia central
    derivada_central = (f(x0 + h) - f(x0 - h)) / (2 * h)
    error_central.append(abs(derivada_central - derivada_exacta))

# Gráfica de los errores
plt.figure(figsize=(10, 6))
plt.loglog(h_values, error_adelante, label="Diferencia Adelante", marker="o")
plt.loglog(h_values, error_atras, label="Diferencia Atrás", marker="x")
plt.loglog(h_values, error_central, label="Diferencia Central", marker="s")
plt.xlabel("Tamaño del paso h")
plt.ylabel("Error absoluto")
plt.title("Error en la aproximación de la derivada con diferencias finitas")
plt.legend()
plt.grid(True, which="both", ls="--", lw=0.5)
plt.show()
```

## Comparacion con el paso Complejo


```python
import numpy as np

def complex_step_derivative(f, x, h=1e-8):
    return np.imag(f(x + 1j * h)) / h

# Example usage:
f = np.sin  # Example function
x_point = np.pi / 4
derivative = complex_step_derivative(f, x_point)
print("Derivative at x =", x_point, "is approximately", derivative)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función compleja para aproximar derivada
def complex_step_derivative(f, x, h=1e-8):
    return np.imag(f(x + 1j * h)) / h

# Parámetros iniciales
f = np.sin              # Función de ejemplo
x_point = np.pi / 4     # Punto de evaluación
exact_derivative = np.cos(x_point)  # Derivada exacta de sin(x) en pi/4

# Valores de h decrecientes
h_values = np.logspace(-1, -15, 100)  # 100 valores de h entre 1e-1 y 1e-15

# Lista para almacenar errores
error_paso_complejo = []

# Cálculo de la derivada y el error para cada valor de h
for h in h_values:
    # Derivada calculada usando el paso complejo
    derivative = complex_step_derivative(f, x_point, h=h)
    # Error absoluto en la aproximación
    error_paso_complejo.append(abs(derivative - exact_derivative))

# Graficar el error
plt.figure(figsize=(10, 6))
plt.loglog(h_values, error_paso_complejo, label="Paso Complejo", marker="^", color="purple")
plt.xlabel("Tamaño del paso h")
plt.ylabel("Error absoluto")
plt.title("Error en la aproximación de la derivada usando el paso complejo")
plt.legend()
plt.grid(True, which="both", ls="--", lw=0.5)
plt.show()
```
