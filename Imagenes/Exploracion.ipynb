### Hagamos otro enfoque para conseguir la matriz dispersa

```python
def matriz_tridiagonal(N, diagonal=-2, off_diagonal=1):
    """
    Genera una matriz tridiagonal de tamao NxN con:
    - `diagonal`: valor en la diagonal principal.
    - `off_diagonal`: valores en las diagonales superiores e inferiores.
    """
    return diags([off_diagonal, diagonal, off_diagonal], [-1, 0, 1], shape=(N, N)).toarray()


Dx = matriz_tridiagonal(N*N, diagonal=-2, off_diagonal=1)

N, M = (3,3)
Dy = diags([1, -2, 1], [-1, 0, 1], shape=(M, M))

A = Dx + kron(Dy, identity(N))
```

```python
# Visualizar la matriz A con cuadrcula y ejes bien definidos
A_dense = A.toarray() if not isinstance(A, np.ndarray) else A

plt.figure(figsize=(6,6))
plt.imshow(A_dense, cmap='viridis', interpolation='none')

# Agregar los valores de la matriz en cada celda
for i in range(A_dense.shape[0]):
    for j in range(A_dense.shape[1]):
        plt.text(j, i, f'{A_dense[i, j]:.1f}', ha='center', va='center', color='white')

# Configurar los ticks en los ejes x e y para que coincidan con las posiciones de los elementos
plt.xticks(np.arange(A_dense.shape[1]))
plt.yticks(np.arange(A_dense.shape[0]))

# Agregar una cuadrcula alineada con los ticks
plt.grid(visible=True, color='black', linewidth=0.5, linestyle='--')

plt.colorbar(label='Valor')
plt.title('Visualizacin de la matriz A con cuadrcula')
plt.xlabel('ndices de columna')
plt.ylabel('ndices de fila')
plt.show()

```
